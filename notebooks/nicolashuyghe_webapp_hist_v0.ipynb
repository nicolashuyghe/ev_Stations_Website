```python
'''Linear algebra'''
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)

'''Data visualisation'''
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.gofplots import qqplot # cf. 5.2
```

# Loading dataset

```python
dataset = pd.read_csv('../raw_data/ytrain_raw.csv')
```

```python
df_static = pd.read_csv('../raw_data/info_static.csv')
```

# Preprocessing

## y_train

```python
# Cleaning
dataset_cleaned = dataset.fillna('Offline')
dataset_cleaned = dataset_cleaned.set_index('timestamp')
dataset_cleaned.head()
```

```python
# Compute nb_charging
for col in dataset_cleaned.columns:
    dataset_cleaned[col] = dataset_cleaned[col].map({'Available':0, 'Charging':1, 'Down':0, 'Offline':0, 'Passive':0})
dataset_cleaned.head()
```

```python
# Rename columns terminals into stations
terminal_names = list(dataset_cleaned.columns)
stations_name = [terminal[:-3] for terminal in terminal_names]
terminals_dict = {terminal_names[i]: stations_name[i] for i in range(len(terminal_names))}
df_stations = dataset_cleaned.rename(columns=terminals_dict)
df_stations.head()
```

```python
# Group terminals by stations
df_stations = df_stations.groupby(level=0,axis=1).sum()
df_stations.head()
```

```python
# Unpivot stations
df_stations = df_stations.melt(value_name='nb_charging',ignore_index=False)
df_stations = df_stations.reset_index().rename(columns={'variable':'s_id'})
df_stations.head()
```

```python
df_stations['timestamp'] = pd.to_datetime(df_stations['timestamp'])
df_stations.info()
```

```python
df_stations['nb_charging'] = df_stations['nb_charging'].astype('int32')
df_stations.info()
```

## Static info

```python
df_static.head()
```

```python
df_static = df_static.groupby(by='s_id').first().reset_index()[['s_id','latitude','longitude','street','city','zipcode']]
df_static.head()
```

# Merge tables

```python
df_merged = df_stations.merge(df_static)
df_merged.head()
```

```python
df_merged.sum(axis=0)
```

# Pydeck visualisation

```python
import pydeck
```

```python
lat, lon = 48.8534, 2.3488

# Define a layer to display on a map
layer = pydeck.Layer(
    'HexagonLayer',
    df_merged,
    get_position=["longitude", "latitude"],
    elevation_scale=50,
    elevation_range=[0, 100],
    extruded=True, 
    pickable=True)

# Set the viewport location
view_state = pydeck.ViewState(
    longitude=lon,
    latitude=lat,
    zoom=6,
    min_zoom=5,
    max_zoom=15,
    pitch=50)
```

```python
r = pydeck.Deck(
    layers=layer,
    initial_view_state=view_state)
```

```python
r.to_html('hexagon-example-nhuyghe.html')
```
